```python
import matplotlib.pyplot as plt
```

```python
import tensorflow as tf
```

```python
set = tf.keras.datasets.mnist.load_data()
```

```python
set
```

```python
(X_train, y_train), (X_test, y_test) = set
```

```python
X_train
```

```python
plt.imshow( X_train[0])
```

```python
from tensorflow.keras import layers, models, datasets
```

```python
X_train.shape
```

```python
X_train = X_train.reshape((60000,28,28,1)).astype('float')/255
```

```python
X_test = X_test.reshape((10000,28,28,1)).astype('float')/255
```

```python
X_train.shape
```

```python
model = models.Sequential([
    layers.Conv2D(32,(3,3), activation='relu', input_shape=(28, 28,1)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (2, 2), activation='relu'),
    layers.MaxPool2D((2, 2)),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics =['accuracy']
             )
```

```python
model.fit(X_train, y_train, epochs=5, batch_size=64)
```

```python
test_loss, test_acc = model.evaluate(X_test, y_test, verbose=2)
print(f'\nPrécision sur le test : {test_acc:.4f}')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Faire des prédictions sur l'ensemble de test
predictions = model.predict(X_test)

def afficher_prediction(index):
    plt.figure(figsize=(2,2))
    
    # Afficher l'image (on la remet en 28x28 pour l'affichage)
    plt.imshow(X_test[index].reshape(28, 28), cmap=plt.cm.binary)
    
    # Récupérer l'index de la probabilité la plus haute (le chiffre deviné)
    prediction_ia = np.argmax(predictions[index])
    label_reel = y_test[index]
    
    color = 'blue' if prediction_ia == label_reel else 'red'
    
    plt.title(f"IA: {prediction_ia} (Réel: {label_reel})", color=color)
    plt.axis('off')
    plt.show()

# Afficher les 5 premières prédictions
for i in range(5):
    afficher_prediction(i)
```
