```python
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_validate

# 1. Chargement des données
data = fetch_california_housing()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target # Prix moyen des maisons (en centaines de milliers $)

# 2. Définition du modèle
# On utilise une Forêt Aléatoire pour capturer des relations complexes
model = RandomForestRegressor(n_estimators=50, random_state=42)

# 3. Choix des métriques pour la validation croisée
metrics = {
    'R2': 'r2',
    'MAE': 'neg_mean_absolute_error',
    'RMSE': 'neg_root_mean_squared_error'
}

# 4. Exécution de la Cross-Validation (cv=5 plis)
results = cross_validate(model, X, y, cv=5, scoring=metrics)

# 5. Affichage des résultats "propres"
print("--- Résultats de l'évaluation ---")
print(f"R² (Score) : {results['test_R2'].mean():.3f} (± {results['test_R2'].std():.3f})")
print(f"MAE (Erreur Moyenne) : {-results['test_MAE'].mean():.3f} k$")
print(f"RMSE (Pénalisation erreurs) : {-results['test_RMSE'].mean():.3f} k$")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris

ir = load_iris()
df = pd.DataFrame(data=ir.data, columns=ir.feature_names)
df

# Chargement d'un dataset d'exemple (Segmentation Client)
#url = "https://raw.githubusercontent.com"
#df = pd.read_csv(url)

# On sélectionne : Revenu Annuel et Score de Dépense
X = df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']]

```

```python

# ÉTAPE CRUCIALE : Normalisation
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Méthode du coude pour trouver K
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss, marker='o')
plt.title('Méthode du Coude')
plt.xlabel('Nombre de clusters')
plt.ylabel('WCSS (Inertie)')
plt.show()
```

```python
# Application du modèle
nb = 4
kmeans = KMeans(n_clusters=nb, init='k-means++', random_state=42)
clusters = kmeans.fit_predict(X_scaled)

# Ajout au DataFrame pour analyse
df['Cluster'] = clusters

# Visualisation graphique
plt.figure(figsize=(10, 6))
for i in range(nb):
    plt.scatter(X_scaled[clusters == i, 0], X_scaled[clusters == i, 1], label=f'Cluster {i}')

plt.title('Segmentation des clients')
plt.xlabel('Revenu Annuel (Normalisé)')
plt.ylabel('Score de Dépense (Normalisé)')
plt.legend()
plt.show()
```
