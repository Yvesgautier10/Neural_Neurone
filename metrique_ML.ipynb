```python
import numpy as np
import pandas as pd
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, f1_score, recall_score
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
# les metriques pour les regressions
y_test = [100, 200, 300]
y_pred = [110, 190, 310]

print(f"Erreur (MSE) : {mean_squared_error(y_test, y_pred)}")
print(f"Score R² : {r2_score(y_test, y_pred)}")
print(f"Erreur (MAE) : {mean_absolute_error(y_test, y_pred)}")
```

```python
# les metriques pour la classification
y_vrai = [0,0,0,0,0,1]
y_pred = [1,0,0,0,1,1]
score = accuracy_score(y_vrai, y_pred)
score
# pourcentage de bonnes reponses
```

```python
# utilisation de la matrice de confusion
matrice = confusion_matrix(y_vrai, y_pred)
matrice
```

```python
matrice1 = recall_score(y_vrai, y_pred)
matrice1
```

```python
print(classification_report(y_vrai, y_pred, target_names=['Classe 0', 'Classe 1']))
```
