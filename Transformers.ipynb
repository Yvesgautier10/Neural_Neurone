```python
# 1. Installation de la bibliothèque leader du marché
!pip install -q transformers torch

from transformers import pipeline

# 2. Chargement d'un modèle de "Sentiment Analysis" pré-entraîné
# Ce modèle a déjà lu des millions de phrases
classifier = pipeline("sentiment-analysis", model="nlptown/bert-base-multilingual-uncased-sentiment")

# 3. Test en français
phrases = [
    "J'adore cette formation, c'est incroyablement clair !",
    "Je suis très déçu, le code ne fonctionne pas sur mon pc.",
    "C'est un cours correct, sans plus."
]

resultats = classifier(phrases)

for i, res in enumerate(resultats):
    print(f"Phrase: {phrases[i]}")
    print(f"Résultat: {res['label']} (Confiance: {res['score']:.2f})\n")
```
